```python
def define_praktikum(row):
    list_p = []
    if row['PS11_01'] == 'ausgewählt' :
        list_p.append('FP A')
    if row['PS11_02'] == 'ausgewählt' :
        list_p.append('FP B')
    if row['PS11_03'] == 'ausgewählt' :
        list_p.append('VP A')
    if row['PS11_04'] == 'ausgewählt' :
        list_p.append('VP B')
    if row['PS11_05'] == 'ausgewählt' :
        list_p.append('EP A')
    return list_p
```

```python
def define_praktikum_one_value(row):
    value=''
    if row['PS11_01'] == 'ausgewählt' :
        value = 'FP A'
    if row['PS11_02'] == 'ausgewählt' :
        value = 'FP B'
    if row['PS11_03'] == 'ausgewählt' :
        value = 'VP A'
    if row['PS11_04'] == 'ausgewählt' :
        value = 'VP B'
    if row['PS11_05'] == 'ausgewählt' :
        value = 'EP A'
    return value
```

```python
def define_praktikum_one_value_w1(row):
    value=''
    if row['D6_01'] == 'ausgewählt' :
        value = 'FP A'
    if row['D6_02'] == 'ausgewählt' :
        value = 'FP B'
    if row['D6_03'] == 'ausgewählt' :
        value = 'VP A'
    if row['D6_04'] == 'ausgewählt' :
        value = 'VP B'
    return value
```

```python
def define_praktikum_one_value_w2(row):
    value=''
    if row['SD23_01'] == 'ausgewählt' :
        value = 'FP A'
    if row['SD23_02'] == 'ausgewählt' :
        value = 'FP B'
    if row['SD23_03'] == 'ausgewählt' :
        value = 'VP A'
    if row['SD23_04'] == 'ausgewählt' :
        value = 'VP B'
    return value
```

```python
def define_praktikum_w8(row):
    list_p = []
    if row['SD08_01'] == 'Praktikum absolviert' :
        list_p.append('EP A')
    if row['SD08_02'] == 'Praktikum absolviert' :
        list_p.append('EP B')
    if row['SD08_03'] == 'Praktikum absolviert' :
        list_p.append('FP A')
    if row['SD08_04'] == 'Praktikum absolviert' :
        list_p.append('FP B')
    if row['SD08_05'] == 'Praktikum absolviert' :
        list_p.append('VP A')
    if row['SD08_06'] == 'Praktikum absolviert' :
        list_p.append('VP B')
    if row['SD08_07'] == 'Praktikum absolviert' :
        list_p.append('MP A')
    return list_p
```

```python
# vals = w3_[['Praktikum_w3','SWE_Mean_norm']].values
wellen = [
    w1_[['Praktikum_w1','SWE_Mean_norm']].values,
    w3_[['Praktikum_w3','SWE_Mean_norm']].values, 
    w4_[['Praktikum_w4','SWE_Mean_norm']].values, 
    w5_[['Praktikum_w5','SWE_Mean_norm']].values, 
    w6_v3_[['Praktikum_w6','SWE_Mean_norm']].values, 
    w8_[['Praktikum_w8','SWE_Mean_norm']].values]

p_set = set()

GR_EP_A = []
GR_EP_B = []
GR_VP_A = []
GR_VP_B = []
GR_FP_A = []
GR_FP_B = []
GR_MP_A = []
for vals in wellen:
    for row in vals:
        p_set.add(row[0])
        if row[0] == 'EP A':
            GR_EP_A.append(row)
        if row[0] == 'EP B':
            GR_EP_B.append(row)
        if row[0] == 'VP A':
            GR_VP_A.append(row)
        if row[0] == 'VP B':
            GR_VP_B.append(row)
        if row[0] == 'FP A':
            GR_FP_A.append(row)
        if row[0] == 'FP B':
            GR_FP_B.append(row)
        if row[0] == 'MP A':
            GR_MP_A.append(row)
        
GR_EP_A = np.array(GR_EP_A)
GR_EP_B = np.array(GR_EP_B)
GR_VP_A = np.array(GR_VP_A)
GR_VP_B = np.array(GR_VP_B)
GR_FP_A = np.array(GR_FP_A)
GR_FP_B = np.array(GR_FP_B)
GR_MP_A = np.array(GR_MP_A)
```

```python
dict = {}
wellen = [
    w1_[['Praktikum_w1','SWE_Mean_norm']],
    w3_[['Praktikum_w3','SWE_Mean_norm']], 
    w4_[['Praktikum_w4','SWE_Mean_norm']], 
    w5_[['Praktikum_w5','SWE_Mean_norm']], 
    w6_v3_[['Praktikum_w6','SWE_Mean_norm']], 
    w8_[['Praktikum_w8','SWE_Mean_norm']]]

wellen_names = ["w1","w3","w4","w5","w6","w8"]
for name,w in zip(wellen_names,wellen):
   dict[name] = {
      "EP A" : w[w[f'Praktikum_{name}'] == 'EP A']["SWE_Mean_norm"].mean(),
      "EP B" : w[w[f'Praktikum_{name}'] == 'EP B']["SWE_Mean_norm"].mean(),
      "FP A" : w[w[f'Praktikum_{name}'] == 'FP A']["SWE_Mean_norm"].mean(),
      "FP B" : w[w[f'Praktikum_{name}'] == 'FP B']["SWE_Mean_norm"].mean(),
      "VP A" : w[w[f'Praktikum_{name}'] == 'VP A']["SWE_Mean_norm"].mean(),
      "VP B" : w[w[f'Praktikum_{name}'] == 'VP B']["SWE_Mean_norm"].mean()
   }
pd.DataFrame.from_dict(dict, orient="index")     
```
